```python
urls = ['https://www.lamoda.ru/p/mp002xw0xkrv/clothes-befree-futbolka/',
        'https://www.lamoda.ru/p/mp002xw041ef/clothes-zarina-futbolka/',
        'https://www.lamoda.ru/p/mp002xw19g9d/clothes-thecave-futbolka/',
        'https://www.lamoda.ru/p/mp002xw19g9d/clothes-thecave-futbolka/',
        'https://www.lamoda.ru/p/ni464emdneu6/clothes-nike-futbolka/',
        'https://www.lamoda.ru/p/to172emdtmo1/clothes-tomtailor-futbolka/']



```

```python
name_text = {}  
review_text = []
review_name = []
for elem in urls:
    page = rq.get(elem)
    soup = BeautifulSoup(page.text, 'html.parser')
    text = soup.find_all(class_="rev__node-text")
    name= soup.find_all(class_="rev__node-head_author")
    
    for entry in text:
        review_text.append(entry.get_text()) #–ø–æ–ª–æ–∂–∏–ª–∏ –≤ —Å–ø–∏—Å–æ–∫ –æ—Ç–∑—ã–≤–æ–≤
    
    for entry in name:
        review_name.append(entry.get_text())#–ø–æ–ª–æ–∂–∏–ª–∏ –≤ —Å–ø–∏—Å–æ–∫ –∏–º–µ–Ω
     
    import numpy as np
    for i in np.arange(len(review_name)):
        name_text[review_name[i]] =  review_text[i]
#print(name_text)
with open('review.txt','w',encoding='utf-8') as out:
    for key,val in name_text.items():
        out.write('{}:{}\n'.format(key,val))
```

```python
import itertools
from razdel import sentenize
from razdel import tokenize as razdel_tokenize
from pymorphy2 import MorphAnalyzer
from collections import Counter, defaultdict
import numpy as np
import re
from string import punctuation
from nltk.corpus import stopwords
```

```python
stops = set(stopwords.words('russian')) # –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∏ —Å–ø–∏—Å–æ–∫ —Å—Ç–æ–ø-—Å–ª–æ–≤
morph = MorphAnalyzer()
```

```python
def normalize(text):
    tokens = re.findall('[–∞-—è—ëa-z0-9]+', text.lower())
    normalized_text = [morph.parse(word)[0].normal_form for word \
                                                            in tokens]
    normalized_text = [word for word in normalized_text if len(word) > 2 and word not in stops]
    
    return normalized_text
```

```python
def preprocess(text):
    sents = sentenize(text) # —Ä–∞–∑–±–∏–ª–∏ —Ç–µ–∫—Å—Ç –Ω–∞ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è
    return [normalize(sent.text) for sent in sents] # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ø–∏—Å–æ–∫ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π
```

```python
normalize(reviews)
```

```python
reviews = preprocess(reviews)
```

```python
def ngrammer(tokens, stops, n=2):
    ngrams = []
    tokens = [token for token in tokens if token not in stops] # –¥–æ–±–∞–≤–ª—è–µ–º –≤—Å–µ —Å–ª–æ–≤–∞ —Ç–µ–∫—Å—Ç–∞ –∫—Ä–æ–º–µ —Å—Ç–æ–ø-—Å–ª–æ–≤
    for i in range(0,len(tokens)-n+1):
        # –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ —Å–ø–∏—Å–∫–∞, –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ, –º–∏–Ω—É—Å –æ–∫–Ω–æ:
        ngrams.append(tuple(tokens[i:i+n])) # –ø—Ä–∏—Å–æ–µ–¥–∏–Ω—è–µ–º —ç–ª–µ–º–µ–Ω—Ç –∏ –≤—Å–µ —Å–æ—Å–µ–¥–Ω–∏–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ä–∞–∑–º–µ—Ä–∞ –æ–∫–Ω–∞
    return ngrams # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ø–∏—Å–æ–∫ —Ç–∞–∫–∏—Ö –∫–æ—Ä—Ç–µ–∂–µ–π
```

```python
word_counter = Counter()

for sent in reviews:
    word_counter.update(ngrammer(sent, n=2, stops=stops))
```

```python
# –∫–∞–∫ –ø–æ–ª—É—á–∏—Ç—å —Å–ø–∏—Å–æ–∫ —Å—Å—ã–ª–æ–∫ —Å –≥–ª–∞–≤–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã

links = [] # –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫, –±—É–¥–µ–º —Å—é–¥–∞ –∫–ª–∞—Å—Ç—å —Å—Å—ã–ª–∫–∏
souplist = soup.find_all('a', {'href' : True}) # –Ω–∞—Ö–æ–¥–∏–º —ç–ª–µ–º–µ–Ω—Ç—ã —Å —ç—Ç–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏, —Å–æ—Ö—Ä–∞–Ω—è–µ–º –≤ —Å–ø–∏—Å–æ–∫
for elem in souplist: # –¥–ª—è —ç–ª–µ–º–µ–Ω—Ç–∞ —Å–ø–∏—Å–∫–∞
    links.append(elem['href']) # –¥–æ–±–∞–≤–ª—è–µ–º –Ω—É–∂–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç(—Å—Å—ã–ª–∫—É) –≤ —Å–ø–∏—Å–æ–∫ –¥–ª—è —Å—Å—ã–ª–æ–∫
```

```python
# –∫–∞–∫ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –¥–∞—Ç—É –Ω–æ–≤–æ—Å—Ç–∏

import datetime as dt
import re

def get_today_date():
    today = dt.date.today()
    return today.strftime("%Y/%b/%d")


todayarticles = [] # —Å–∫–ª–∞–¥—ã–≤–∞–µ–º —Å—é–¥–∞ —Å—Å—ã–ª–∫–∏ –Ω–∞ —Å—Ç–∞—Ç—å–∏ –∑–∞ —Å–µ–≥–æ–¥–Ω—è
for link in links: # –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç–∞—Ç—å–∏ –≤ links
    if re.search(get_today_date(), link) != None: # –µ—Å–ª–∏ —É–∫–∞–∑–∞–Ω–∞ –¥–∞—Ç–∞
        todayarticles.append(link) # –¥–æ–±–∞–≤—å—Ç–µ –≤ —Å–ø–∏—Å–æ–∫ —Å—Ç–∞—Ç–µ–π
```
